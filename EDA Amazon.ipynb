```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
url = 'https://drive.google.com/uc?id=1yKlW0x5yyRQijmWbEg5YCsq4wqLl3_oB'
df = pd.read_csv(url)
df
```

```python
df.columns
```

```python
df.rating.unique()
```

```python
df['rating'] = df.rating.replace('|',np.nan)
df['rating'] = pd.to_numeric(df.rating,errors='coerce')
```

```python
df['rating'] = df['rating'].fillna(df.rating.mean())
```

```python
#What is the average rating for each product category
df.groupby('category')['rating'].mean()
```

```python
#What are the top rating_count products by category
df.sort_values(['category', 'rating_count'], ascending=[True, False]).groupby('category').head(3)[['category', 'product_name', 'rating_count']]
```

```python
#What is the distribution of discounted prices vs. actual prices
df['actual_price']=df['actual_price'].str.replace("₹","").str.replace(",","")
```

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹","").str.replace(",","")
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'])
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
```

```python
df[['discounted_price', 'actual_price']].describe()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['discounted_price', 'actual_price']])
plt.title("Distribution of Discounted vs Actual Prices")
plt.ylabel("Price (₹)")
plt.show()
```

```python
# What are the most popular product keywords
from collections import Counter
import re

# join all product names into one big string
text = ' '.join(df['product_name'].astype(str))

# convert to lowercase and split into words (remove punctuation)
words = re.findall(r'\b[a-zA-Z]{3,}\b', text.lower())

# count the most common words
Counter(words).most_common(20)
```

The keyword frequency analysis of product names shows that terms like “USB” (417), “Cable” (414), “Charging” (247), “Fast” (212), and “Smart” (175) are the most common product keywords.
This indicates that the dataset is dominated by electronics, charging accessories, and tech gadgets.
Additional keywords such as “Bluetooth”, “Wireless”, and “LED” further highlight a trend toward smart and connected devices, while terms like “Heater” and “Water” suggest the presence of home and utility products.

```python
# What are the most popular product reviews
df['review_title'].value_counts().head(10)
```

The analysis of product review titles shows that the most frequently used phrases are “Good product”, “Value for money”, “Satisfied”, and “Worth the money.”

Some detailed reviews such as “Worked on iPhone 7 and didn’t work on XR” also appear multiple times, indicating users often share device compatibility or specific product behavior.

```python
#What is the correlation between discounted_price and rating
df[['discounted_price','rating']].corr()
```

```python

plt.scatter(df['discounted_price'], df['rating'], alpha=0.5)
plt.title('Correlation between Discounted Price and Rating')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Rating')
plt.show()
```

The correlation between discounted_price and rating is approximately +0.12, indicating a very weak positive relationship.
This suggests that higher-priced products receive slightly higher ratings, but the effect is minimal.
Overall, customer ratings are not strongly influenced by the discounted price — product quality and performance likely play a more significant role in determining customer satisfaction.

# Identify any potential areas for improvement or optimization based on the data analysis.
Based on the analysis, several improvement opportunities were identified.
Products in lower-rated categories (e.g., Office and Stationery) may benefit from enhanced quality control.
The weak correlation between price and ratings indicates that customers value performance over price, suggesting room for smarter discount strategies.
Increasing the number of customer reviews can improve trust and visibility.
Categories such as Electronics and Computer Accessories, which already perform well, should be prioritized for product expansion.
Lastly, improving clarity in product descriptions and ensuring better compatibility can further enhance customer satisfaction and reduce negative feedback.
