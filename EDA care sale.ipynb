```python
#What is the distribution of car prices for each transmission type
#and how do the interquartile ranges compare
df.groupby("Transmission")['Price ($)'].describe()
```

```python
iqr = df.groupby('Transmission')['Price ($)'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))
iqr
```

The distribution of car prices differs slightly between transmission types.

Comparing the interquartile ranges (IQR), manual cars have a slightly higher IQR ($17,000) compared to automatic cars ($16,500). This indicates that manual car prices are a bit more spread out in the middle 50%, while automatic car prices are slightly more consistent.

```python
# How does car price vary with engine size across different car colors, and which colors have the highest price variation
df.groupby(['Engine','Color'])['Price ($)'].mean()
```

```python
df.groupby('Color')['Price ($)'].std()
```

The average selling price of cars varies across different combinations of engine type and color. Among Double Overhead Camshaft (DOHC) cars, Pale White models have the highest average price (₹30,348), followed by Red and Black variants. For Overhead Camshaft (OHC) engines, Black cars show the highest mean price (₹31,733), while Pale White cars show the lowest (₹24,328).

This indicates that both engine type and color together influence car pricing, with certain color–engine combinations (such as OHC Black) positioned in higher price segments.

When analyzing price variation using standard deviation, Red cars exhibit the highest price variability (std ≈ 15,519), followed closely by Black cars. Pale White cars have the lowest variation, suggesting more consistent pricing within that color category.

Overall, Red-colored vehicles show the widest price spread, while Pale White vehicles are priced more uniformly.

```python
#Is there any seasonal trend in car sales based on the date of sale
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].dt.month
df['month'].value_counts().sort_index()
```

A clear seasonal trend is observed in the monthly car sales data. Sales are lowest in January (790) and February (735), indicating a slow start to the year. From March to August, sales gradually increase and remain relatively stable, ranging between 1500 and 1900 units.

A sharp rise in sales is seen during September (3305), followed by even higher peaks in November (3470) and December (3546), which are the highest sales months of the year.

This pattern suggests a strong seasonal effect, with significantly higher car purchases occurring during the festival season and year-end period, likely driven by festive discounts, promotions, and consumer buying behavior. Overall, the data shows a clear upward seasonal trend in the last quarter of the year.
